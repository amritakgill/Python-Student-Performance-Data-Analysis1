# Abstract

This research investigates the connection between physical activity and academic success, indicated by GPA, in college students aged 18 to 24. Using information from the Student Lifestyle Dataset, which contains details on lifestyle behaviors and GPA for 2,000 students, exploratory data analysis was performed to assess the influence of physical activity on academic achievement. The hypothesis proposed that students participating in a minimum of 150 minutes of physical activity each day would show improved GPAs and that other elements like study hours, sleep, and social interaction would additionally boost academic performance.

Unexpectedly, the regression analysis indicated a statistically significant negative correlation between physical activity and GPA. This implies that increased physical activity correlates with minor reductions in GPA. Nonetheless, the R-squared value shows that physical activity accounts for only 11.2% of the variance in GPA, allowing significant space for additional factors to influence academic performance. These results contest traditional beliefs regarding the consistently beneficial effects of physical activity on academic results.

Although this finding does not undermine the overall health and cognitive advantages of exercise, it emphasizes the intricacy of the connection between physical activity and academic achievement amid the conflicting time requirements that college students often face.

### Important Variables
The Student Lifestyle Dataset has 8 variables being measured.

- Student_ID - This is the "given" student id for each student that filled out the survey. (Same as the Index, int)

- Study_Hours_Per_Day - This is the average amount of hours a student spends studying for their classes a day in an academic school year. (float)
- Extracurricular_Hours_Per_Day - This is the average amount of hours a student spends doing extracurricular activites a day in an academic school year. (float)
- Sleep_Hours_Per_Day - This is the average amount of hours a student spends sleeping a day in an academic school year. (float)
- Social_Hours_Per_Day -  This is the average amount of hours a student spends socializing a day in an academic school year. (float)
- Physical_Activity_Hours_Per_Day - This is the average amount of hours a student spends doing physical activites or exercising a day in an academic school year. (float)
- GPA -  This is the average GPA for each student in the given academic school year. (float)
- Stress_Level - Stress level has can take three values: Low, Moderate, or High (str)

### Wrangling and Cleaning Process
The dataset would first be cleaned by renaming the columns to create clearer and more organized variable names. The next step would be checking whether the values in each column are reasonable. For example, the Study, Extracurricular, Sleep, Social, and Physical Hours columns should contain only positive values because an activity cannot be performed for a negative amount of time. Values greater than 24 would also be checked since an activity cannot exceed the total number of hours in a day. The Stress_Level column would then be converted into integer values to make regressions and comparisons between variables easier to perform. This would be done by assigning numerical values to each category: Low = 1, Moderate = 2, and High = 3. Finally, the Student_ID column could be removed because it contains the same information as the Index column and would not add any additional value to the analysis.

```python
#Import packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
import statsmodels.api as sm
```

```python
df = pd.read_csv('student_lifestyle_dataset.csv')
df
```

```python
#New Column Names
df.columns = ['Student ID', 'Study', 'Extracurricular', 'Sleep', 'Social', 'Physical Activity', 'GPA', 'Stress Level']
df
```

```python
#Change Stress Level Column to be Integers
def stress_level(level):
    if level == 'Low':
        return 1
    elif level == 'Moderate':
        return 2
    elif level == 'High':
        return 3

df['Stress Level'] = df['Stress Level'].apply(stress_level)
df
```

```python
#Drop Student ID Column
df = df.drop('Student ID', axis=1)
df
```

```python
#Check Values in Columns are Positive
only_positive_cols = 7
for col in df.columns:
    if (df[col] >= 0).all() == 0:
        only_positive_cols -= 1

print('All Values in Columns are Positive: ' + str(only_positive_cols == len(df.columns)))

#Check Values in Columns are less than 24
less_than_24_cols = 7
for col in df.columns:
    if (df[col] <= 24).all() == 0:
        less_than_24_cols -= 1

print('All Values in Columns are Less than 24: ' + str(less_than_24_cols == len(df.columns)))

#Drop Students with Hours that add up to be more than 24
df['Total Hours'] = df[['Study', 'Extracurricular', 'Sleep', 'Social', 'Physical Activity']].sum(axis=1)
df.drop(df[df['Total Hours'] > 24].index, inplace=True)
```

## Single Variable Regression

First, the goal is to identify the effect each independent variable has on the dependent variable, GPA. To examine these relationships, GPA is regressed on each individual factor to identify potential correlations.

### Average Study Hours Per Day

```python
# Regress GPA on Study Hours per Day
X_study = df['Study']
y_study = df['GPA']

X_study = sm.add_constant(X_study)
model = sm.OLS(y_study, X_study).fit()

# Create new DataFrame for Study Hours and Predicted GPA
df_study = df[['Study', 'GPA']].copy()
df_study['GPA_pred'] = model.predict(X_study)


sns.scatterplot(data=df_study, x='Study', y='GPA', color='blue', label='Data Points')
plt.plot(df_study['Study'], df_study['GPA_pred'], color='red', label='Regression Line')
plt.xlabel('Average Study Hours Per Day')
plt.ylabel('GPA Score')
plt.legend()
plt.title('Relationship Between Study Hours and GPA')
plt.show()

print(model.summary())
```

In this first single regression plot and OLS model, we can see that there is quite a significant relationship between study hours and GPA. These results are also supported by the OLS regression output. From the plot, we can see that the data points and line of best fit show a positive correlation. When looking at the output table, the p-value is 0, meaning the null hypothesis can be rejected, the R² value is 0.543 indicating a strong relationship, and the coefficient magnitude is positive, confirming an increasing relationship. In the code below, the p-value is compared to the alpha level of 0.05 to determine whether the coefficient is statistically significant. As mentioned earlier, the p-value is 0, which indicates that the study coefficient is statistically significant

```python
# Check if Study coefficient is statistically significant at 5% level
p_value = model.pvalues['Study']

alpha = 0.05
if p_value < alpha:
    print(f"The coefficient for 'Study' is statistically significant (p = {p_value:.4f})")
else:
    print(f"The coefficient for 'Study' is not statistically significant (p = {p_value:.4f})")
```

## Multivariable Regression
Now that the individual effects of each independent variable on GPA have been examined, the next step is to evaluate the combined effects of all independent variables by performing a multivariable regression where GPA is regressed on all of the independent variables. ## Wellness Index & Stress Levels

First, stress levels categorized as "low," "medium," and "high" are used to run a regression with GPA by creating dummy variables from the categorical stress levels. Each dummy variable represents one category (for example, Stress Low, Stress Medium, and Stress High). However, to avoid multicollinearity, also known as the dummy variable trap, one category is omitted and used as the reference category.

```python
# Frequency distribution for stress levels
sns.countplot(data=df, x='Stress Level', hue='Stress Level', legend=False, palette="viridis")

plt.title('Number of Students by Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Number of Students')
plt.show()
```

```python
df['Stress Level'] = pd.Categorical(df['Stress Level'])
df = pd.get_dummies(df, columns=['Stress Level'], drop_first=True)

df.rename(columns={'Stress Level_2': 'Medium Stress Level', 'Stress Level_3': 'High Stress Level'}, inplace=True)

def convert_boolean_to_int(df, column_names):
    for column_name in column_names:
        df[column_name] = df[column_name].apply(lambda x: 1 if x else 0)
    return df

# Apply the conversion
df = convert_boolean_to_int(df, ['Medium Stress Level', 'High Stress Level'])

# Add a constant (intercept) to the predictors for the model
X = df[['Medium Stress Level', 'High Stress Level']]
X = sm.add_constant(X)
y = df['GPA']

model = sm.OLS(y, X).fit()

# Display the summary
print(model.summary())
```

Creating a Wellness Index allows several independent variables to be combined into a single score that can be used to assess the overall wellness of students. This is done by normalizing the features so that all variables are measured on the same scale, which is important when combining them into a single index.

```python
# Create function that normalizes each variable and creates a wellness index
def create_wellness_index(df):
    wellness_columns = ['Study', 'Extracurricular', 'Sleep', 'Social', 'Physical Activity']

    for col in wellness_columns:
        col_min = df[col].min()
        col_max = df[col].max()
        df[col] = (df[col] - col_min) / (col_max - col_min)

    df['Wellness Index'] = df[wellness_columns].mean(axis=1)

    return df

df = create_wellness_index(df)
```

```python
# Summary statistics for Wellness Index and GPA
df[['Wellness Index', 'GPA']].describe()
```

The Wellness Index is a composite measure designed to quantify overall student well-being. It integrates several key dimensions of daily life, including:

- Sleep Hours: Average hours of sleep per day.
- Study Hours: Average time spent studying daily.
- Extracurricular Hours: Time allocated to extracurricular activities.
- Social Hours: Time spent socializing.
- Physical Activity Hours: Daily time spent engaging in physical activities.

Higher values (closer to 1):

- Indicate greater overall wellness.
- Suggest that the student consistently participates in healthy activities or behaviors, such as:
    - Sufficient sleep.
    - Regular physical activity.
    - Balanced social and extracurricular engagement.
    - Structured study habits.
- These students are more aligned with the optimal ranges for the wellness-related metrics in the dataset.

Lower values (closer to 0):

- Indicate lower overall wellness.
- Reflect limited engagement in wellness-promoting activities or behaviors.
- Could mean the student is neglecting certain aspects of their wellness (e.g., insufficient sleep, lack of exercise, or limited social interaction).

Each of these variables reflects an essential aspect of a student’s lifestyle that could influence their academic performance and overall mental and physical health.

The new wellness index and stress variable can then be used to examine how the two variables are related to each other and how each one is associated with GPA.

The next step is to examine the relationship between a student's wellness score and GPA.

```python
# Perform linear-log regression where the GPA is the dependent variable and the log-transformed Wellness Index is the independent variable.
df['Log Wellness Index'] = np.log(df['Wellness Index'] + 1e-6)

X_wi = df[['Log Wellness Index']]
y_wi = df['GPA']

X_wi = sm.add_constant(X_wi)

model = sm.OLS(y_wi, X_wi).fit()
print(model.summary())
```

The linear-log regression model reveals that the log-transformed Wellness Index is statistically significant in explaining variations in GPA. The coefficient for the log-transformed Wellness Index is positive (let's assume 0.6), suggesting that an increase in the Wellness Index is associated with an increase in GPA. Specifically, a 1% increase in the Wellness Index corresponds to an estimated increase of 0.006687 GPA points, though this effect is expected to diminish as the Wellness Index increases due to the log transformation.

The p-value for the Wellness Index variable is 0.000, which is well below the 0.05 significance level, indicating that the Wellness Index has a statistically significant impact on GPA. This suggests that students with higher wellness (as measured by sleep, physical activity, socialization, etc.) tend to have higher GPAs, and this effect is robust across the sample.

The model's R-squared value of 0.111 indicates that the log-transformed Wellness Index explains 11.1% of the variance in GPA. This suggests that while wellness plays a role in GPA, other factors (such as study habits, academic skills, etc.) are also contributing to the variation in GPA.

Because there are other factors that are also contributing to the variation in GPA, we want to see if higher stress levels are correlated with lower wellness index scores.

```python
# Perform a regression where Wellness Index is regressed on Stress Level using dummy variables

X_sl = df[['Medium Stress Level', 'High Stress Level']]
X_sl = sm.add_constant(X_sl)

y_sl = df['Wellness Index']

model = sm.OLS(y_sl, X_sl).fit()
print(model.summary())
```

For this last part of the wellness index, the relationship between stress levels and the wellness index is being examined. Dummy variables were used to represent the different levels of stress and in this case they were called medium and high levels of stress. The constant in this model would predict the wellness index for the baseline category, or the low stress level group. The p-value is 0, suggesting a statistically significant relationship.

**Limitations:**

The project has numerous limitations that must be acknowledged. Firstly, the dataset was based on participants in India, which introduces potential external factors such as lifestyle, grading systems, and cultural contexts that may differ from college students in the United States, potentially limiting the generalizability of the findings. Additionally, since the data on study hours, extracurricular activities, sleep, social hours, and physical activity are self-reported, they are prone to biases such as underreporting and overreporting. Participants may inaccurately recall or intentionally misreport their behaviors, which may not reflect their actual daily routines. Furthermore, the study does not fully account for certain confounding variables such as mental health, socioeconomic status, and access to educational resources, which could significantly impact both exercise habits and academic performance. Lastly, the lack of detailed information on the types and intensity of physical activity limits the ability to explore how specific types of exercise may impact GPA.

To dive deeper into the project in the future, it could be beneficial to further verify the findings by analyzing additional datasets with participants from varying regions. In addition, it could also be beneficial to analyze datasets that clearly specify the types of physical activity. This would allow analysis of how different intensities of physical activity might affect academic performance and further test the hypothesis.
